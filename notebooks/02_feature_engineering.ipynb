```python
prediction_file_path = "../data/raw/prediction_input.parquet"
input_dataset_file_path = "../data/raw/input_dataset-2.parquet"
bolt_pretension_file_path = "../data/raw/bolt_pretension.csv"

data_df = pd.read_parquet(input_dataset_file_path)
pretension_df = pd.read_csv(bolt_pretension_file_path, header=None)
prediction_df = pd.read_parquet(prediction_file_path)


pretension_df.drop(pretension_df.columns[[0]], axis=1, inplace=True)
print(pretension_df)
```

# Adjusted tensile

```python
tensile_cols = [c for c in data_df if "Tensile" in c]
tensile_adjusted = data_df[tensile_cols] - pretension_df[1].values
tensile_adjusted.rename(columns=lambda x: str(x) + "_adj", inplace=True)
tensile_adjusted.describe()
```
